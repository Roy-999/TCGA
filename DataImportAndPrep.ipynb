# Collate case files from directories into one dataframe

```python
import os
import pandas as pd

# Define the root directory containing all the folders
root_dir = 'C:\Lappy\Swapnil\Business Analytics\Dissertation\Data'

# Initialize an empty list to store DataFrames
dataframes = []
file_names= []
counter = 0
# Loop over all folders and files in the root directory
for subdir, _, files in os.walk(root_dir):
    for file in files:
        # Create the full file path
        file_path = os.path.join(subdir, file)
        if file[-3:]=='txt':
            file_names.append(file)
        try:
            # Read the file into a DataFrame (assuming CSV format, modify as needed)
            if counter == 0:
                df = pd.read_csv(file_path, sep = '\t', header=None)
                counter+=1
            else:
                df = pd.read_csv(file_path, sep = '\t', header=None, usecols=[1])
            # Append the DataFrame to the list
            dataframes.append(df)
        except Exception as e:
            #print(f"Error reading {file_path}: {e}")
            pass

# Now 'dataframes' contains all the DataFrames read from the files
```

```python
file_names = ['composite'] + file_names
```

```python
new_df = pd.concat(dataframes, axis=1)
```

```python
new_df.columns = file_names
```
